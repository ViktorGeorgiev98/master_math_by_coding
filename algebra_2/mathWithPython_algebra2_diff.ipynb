```python
v = np.arange(0,11)
print(v)
print(np.diff(v))
print(np.diff(v,2))
print(np.diff(np.diff(v)))
```

```python
x = np.linspace(-2, 2, 1000)
f = x**2
df = np.diff(f)
dx = np.diff(x)
plt.plot(x, f, label="f")
plt.plot(x[0:-1], df/dx, "r", label="df/dx")
plt.axis([-2,2,-1,2])
plt.grid()
plt.legend()
plt.show()
```
